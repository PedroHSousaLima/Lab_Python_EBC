```python
import pyautogui as pag
import time
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service

pag.PAUSE = 2

servico = Service(ChromeDriverManager().install())

navegador = webdriver.Chrome(service=servico)

# Lista de Cidades

cidades = [
    #'PLANOPILOTO',
    #'CRUZEIRO',
    'OCTOGONAL',
    'LAGONORTE',
    'LAGOSUL',
    'SUDOESTE',
    'PARANOA',
    'RIACHOFUNDO',
    'NÚCLEOBANDEIRANTES',
    'GUARA',
    'CANDAGOLÂNDIA',
    'SÃOSEBASTIÃO',
    'GAMA',
    'RECANTODASEMAS',
    'SANTAMARIA',
    'TAGUATINGA',
    'ÁGUASCLARAS',
    'CEILÂNDIA',
    'SAMAMBAIA',
    'BRAZLÂNDIA',
    'SOBRADINHO',
    'PLANALTINA'
]

for bairro in cidades:
    navegador.get('https://www.google.com.br/maps')
    navegador.find_element('xpath', '//*[@id="searchboxinput"]').send_keys(bairro)
    navegador.find_element('xpath', '//*[@id="searchbox-searchbutton"]').click()
    time.sleep(3)
#Copiar a URL
    pag.hotkey('ctrl','l')
    time.sleep(3)
    pag.hotkey('ctrl','c')
    time.sleep(3)
    pag.click(x=1205, y=9, clicks=1)
    time.sleep(3)
    pag.press('F2')
    time.sleep(3)
    pag.hotkey('ctrl','v')
    time.sleep(3)
    pag.press('enter')
    time.sleep(3)
    pag.hotkey('alt','tab')
    
    

```
